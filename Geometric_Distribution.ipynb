```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

p_value = [0.2, 0.4, 0.6, 0.8]
colors = ['red', 'blue', 'orange', 'green']
x = np.arange(1, 11)

for p, color in zip(p_value, colors):
    pmf = geom.pmf(x, p)
    plt.plot(x, pmf, 'o-', color = color, label = f'p = {p}')
    
plt.xlabel('x')
plt.ylabel('p')
plt.xticks(np.arange(1, 11))


plt.legend()
plt.show()
```
